## Correlation matrix

```python
# Sélectionner les colonnes numériques
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

# Calculer le coefficient de corrélation r entre "price" et les colonnes numériques
corr_matrix = df[numeric_cols].corr()["price"]

# Afficher le coefficient de corrélation r pour chaque colonne numérique
print(corr_matrix)
```

carheight, stroke, compressionratio, peakrpm : pas de corrélation significative

```python
# drop carname variable
df = df.drop('carheight', axis=1)
df = df.drop('stroke', axis=1)
df = df.drop('peakrpm', axis=1)
df = df.drop('compressionratio', axis=1)
```
